```python
# Create boxplot with two variables: 
fig, ax = plt.subplots(figsize= (14,8))
ax.set_ylabel('Profit')
boxplot = df[['Profit', 'Sales']].boxplot(by='Profit', figsize=(13,8), 
                                            vert=False, sym='b.', ax=ax)
  
```

```python
# Create scatter plot with two variables
plt.figure(figsize=(14,8))
scatterplot = sns.regplot(data=df, x='Profit', y='Sales')
```

```python
# Create tabular correlation matrix
corr = df.corr()
heatmap, ax = plt.subplots(figsize=(13,10))
# Graph correlation matrix
heatmap = sns.heatmap(corr, ax=ax, xticklabels=corr.columns.values, 
                      yticklabels=corr.columns.values, cmap = 'coolwarm' )
```
